```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('80_no_correct_B_wall_0.25_final (2).png')


grid_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,8))
plt.imshow(grid_RGB) # Printing the original picture after converting to RGB


grid_HSV = cv2.cvtColor(grid_RGB, cv2.COLOR_RGB2HSV) # Converting to HSV
lower1_red = np.array([0, 40, 40])
upper1_red = np.array([1, 255, 255])


lower2_red = np.array([160,40,40])
upper2_red = np.array([179,255,255])

lower_mask = cv2.inRange(grid_HSV, lower1_red, upper1_red)
upper_mask = cv2.inRange(grid_HSV, lower2_red, upper2_red)

mask=lower_mask + upper_mask;

res = cv2.bitwise_and(img, img, mask=mask) # Generating image with the green part

print("Red Part of Image")
plt.figure(figsize=(20,8))
plt.imshow(res)

RED_perc = (mask>0).mean()
print(RED_perc)
```
